# ls 명령어 설명
ls 명령어는 유닉스 계열 운영 체제(Linux 등)에서 현재 디렉토리 또는 지정한 디렉토리 내의 파일과 디렉토리 목록을 표시하는 필수 명령어입니다.

list segments의 약자로, 파일명, 디렉토리 구조, 그리고 -l 옵션을 통한 권한, 소유자, 크기 등 상세 정보를 확인하는 데 사용됩니다.

# 자주 사용하는 ls 옵션
ls -l: 파일의 상세 정보(권한, 소유자, 크기, 수정 시간)를 나열합니다.

ls -a: 숨김 파일(점.으로 시작하는 파일)을 포함한 모든 파일을 표시합니다.

ls -al: 상세 정보와 함께 숨김 파일을 모두 보여줍니다.

ls -h: -l 옵션과 함께 사용하여 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시합니다.

ls -t: 파일을 수정 시간 순서대로 최신순으로 정렬하여 표시합니다.

ls -r: 목록을 역순으로 정렬합니다.

ls -R: 하위 디렉토리의 내용까지 재귀적으로 모두 나열합니다.

ls -d: 디렉토리 자체의 정보를 보여줍니다.

# 활용 예시
ls -alF: 모든 파일을 상세 정보와 함께 표시하고, 파일 형식을 알리는 기호(*, / 등)를 끝에 붙입니다.

ls -ltR: 하위 디렉토리를 포함하여 시간순으로 정렬된 목록을 상세히 출력합니다.

# CD 명령어
리눅스의 cd(Change Directory) 명령어는 현재 작업 중인 디렉토리(폴더)를 이동할 때 사용하는 가장 기초적인 명령어입니다.

절대 경로 또는 상대 경로를 사용하여 디렉토리를 변경할 수 있으며,

cd ..으로 상위 폴더로, cd 또는 cd ~으로 홈 디렉토리로 바로 이동할 수 있습니다.
# 주요 cd 명령어 사용법:
cd [디렉토리명]: 현재 위치 내의 특정 디렉토리로 이동

cd ..: 상위 디렉토리로 한 단계 이동

cd /: 루트(/) 디렉토리로 이동

cd ~ 또는 cd: 현재 사용자의 홈 디렉토리로 이동

cd -: 바로 이전에 위치했던 디렉토리로 이동

cd /home/user/docs: 절대 경로를 사용하여 특정 디렉토리로 직접 이동

팁:
디렉토리 이름을 입력할 때 Tab 키를 사용하면 자동 완성을 통해 더 편하게 이동할 수 있습니다.

현재 위치는 pwd 명령어로 확인 가능합니다.

# touch 명령어
리눅스의 touch 명령어는 0바이트 빈 파일을 생성하거나,

기존 파일의 최근 접근/수정 시간(atime, mtime)을 현재 시간으로 갱신하는 데 사용됩니다.

파일이 없으면 새로 만들고, 있으면 시간만 변경하며 touch [파일명] 형태로 사용합니다.

# 주요 기능 및 사용법
빈 파일 생성: touch newfile.txt (파일이 없으면 생성)

시간 갱신: touch existingfile.txt (이미 존재하는 파일의 시간 정보를 현재로 업데이트)

여러 파일 생성: touch file1.txt file2.txt (공백으로 구분하여 동시에 여러 파일 생성)

# 주요 옵션
-c: 파일이 없어도 새로 생성하지 않습니다. (시간 업데이트만 수행)

-a: 최근 접근 시간(access time)만 변경합니다.

-m: 최근 변경 시간(modification time)만 변경합니다.

-t: [[CC]YY]MMDDhhmm[.ss] 형식으로 시간을 직접 지정하여 변경합니다.

# rm 명령어
리눅스의 rm (remove) 명령어는 파일이나 디렉토리를 삭제하는 핵심 명령어입니다.

기본적으로 파일을 삭제하며, -r (디렉토리 삭제), -f (강제 삭제) 등의 옵션을 사용하여

하위 디렉토리를 포함한 폴더 전체를 강제로 지울 수 있습니다.

삭제된 파일은 복구가 어렵기 때문에 주의가 필요합니다.
# 주요 사용법 및 옵션
파일 삭제: rm [파일명] (기본적으로 삭제 여부 확인)

-i (Interactively): 삭제 전 물어봄 (rm -i file.txt)

-r (Recursive): 디렉토리 및 내부 파일 전체 삭제 (rm -r folder)

-f (Force): 존재하지 않는 파일은 무시하고, 묻지 않고 강제 삭제 (rm -f file.txt)

-rf: 디렉토리를 묻지 않고 강제로 삭제 (rm -rf folder)

# cp 명령어
리눅스 cp (copy) 명령어는 파일이나 디렉터리를 복사할 때 사용하며, 기본 구문은 cp [옵션] [원본] [대상]입니다.

파일 복사, 이름 변경 복사, 디렉터리 복사(-r), 덮어쓰기 확인(-i), 권한 유지(-p) 등

다양한 옵션을 지원하여 시스템 관리에서 필수적인 명령어입니다.
# 주요 cp 명령어 사용법 및 예제
파일 복사: cp file1.txt file1_bak.txt (file1을 file1_bak로 복사)

파일을 디렉터리로 복사: cp file1.txt /home/user/documents/

디렉터리 복사 (하위 포함): cp -r dir1 dir2 (-r 옵션 필수)

복사 시 덮어쓰기 확인: cp -i file1.txt file2.txt (대상 파일 존재 시 물어봄)

파일 속성(권한) 유지하며 복사: cp -p file1.txt file1_bak.txt

여러 파일 복사: cp file1.txt file2.txt /dest/dir/

# 자주 사용하는 핵심 옵션
-r, -R: 디렉터리를 복사할 때 하위 디렉터리와 파일을 모두 복사

-i: 복사할 목적지에 동일한 이름의 파일이 존재할 경우, 덮어쓸지 물어봄

-f: 목적지에 파일이 있어도 강제로 덮어씀

-p: 원본 파일의 권한, 소유자, 시간 등 속성을 보존

-v: 복사 과정을 화면에 표시 (verbose)

# mv 명령어
리눅스 mv (move) 명령어는 파일이나 디렉토리를 다른 위치로 이동하거나 이름을 변경할 때 사용합니다.

기본 문법은 mv [옵션] <원본> <대상>이며, 같은 디렉토리 내에서 대상을 다르게 지정하면 이름이 변경되고,

다른 디렉토리 경로를 지정하면 파일이 이동됩니다.
# 주요 활용법 및 명령어:
파일 이름 변경: mv old_name.txt new_name.txt

파일 위치 이동: mv file.txt /path/to/destination/

디렉토리 이동: mv /source_dir /path/to/destination/

여러 파일 이동: mv file1.txt file2.txt /dest_dir/

# 주요 옵션:
-i (interactive): 덮어쓰기 전 확인 메시지를 표시

-f (force): 덮어쓰기 확인 없이 강제 이동

-n (no-clobber): 대상 파일이 존재하면 덮어쓰지 않음

-v (verbose): 이동 중인 파일 목록을 상세히 출력

# 주의사항:
동일한 이름의 파일이 목적지에 있으면 경고 없이 덮어씌워지므로 주의가 필요합니다.

-i 옵션을 사용하여 덮어쓰기 전에 확인하는 것이 안전합니다.

# mkdir 명령어
리눅스 mkdir(make directory)은 새로운 디렉토리(폴더)를 생성하는 명령어입니다.

mkdir [옵션] 디렉토리명 형식을 사용하며, -p 옵션으로 상위 디렉토리까지 한번에 생성하거나, -m으로 권한을 설정할 수 있습니다.

# 주요 사용법 및 옵션
기본 생성: mkdir folder1 (현재 위치에 folder1 생성)

여러 디렉토리 생성: mkdir dir1 dir2 dir3 (공백으로 구분)

하위 디렉토리 포함 생성 (-p): mkdir -p parent/child (상위 디렉토리 없어도 자동 생성)

권한 지정하여 생성 (-m): mkdir -m 755 new_folder (생성과 동시에 권한 설정)

생성 내용 출력 (-v): mkdir -v new_folder (디렉토리 생성 메시지 출력)

# 주의사항:
동일한 이름의 디렉토리가 있으면 에러가 발생하므로, 기존 디렉토리가 있어도 에러 없이 진행하려면 -p 옵션을 활용합니다.

# cat 명령어
리눅스의 cat(concatenate) 명령어는 파일 내용을 터미널에 출력하거나,

여러 파일을 합치고(연결), 새로운 파일을 생성하는 가장 기본적인 텍스트 처리 도구입니다.

주요 사용법은 cat [파일]로 내용을 확인하거나, > 연산자와 함께 사용하여 파일 생성/저장에 활용합니다.

# 주요 기능 및 사용 예제
파일 내용 출력: cat file.txt (file.txt의 내용을 화면에 표시)

파일 내용 번호와 함께 출력: cat -n file.txt (-n 옵션 사용)

여러 파일 합치기: cat file1.txt file2.txt > combined.txt (두 파일을 합쳐서 새로운 파일로 저장)

파일 내용 추가하기: cat file1.txt >> file2.txt (file1 내용을 file2 끝에 덧붙임)

파일 생성: cat > newfile.txt (내용 입력 후 Ctrl+D로 저장 및 종료)

탭/줄 끝 표시: cat -T file.txt (탭 문자), cat -E file.txt (줄 끝)

cat은 작은 크기의 파일 내용을 빠르게 확인할 때 매우 유용하며,

grep이나 more와 같은 다른 명령어와 파이프(|)로 연결하여 자주 사용합니다.

# file 명령어
리눅스 file 명령어는 파일의 확장자나 이름에 의존하지 않고,

파일의 내용(헤더/매직넘버)을 분석하여 정확한 파일 형식(ASCII 텍스트, 바이너리, 이미지 등)을 식별하는 도구입니다.

file [옵션] 파일명 형식으로 사용하며, 주요 옵션으로 -b(유형만 표시), -i(MIME 타입 표시)가 있습니다.

# 주요 활용법 및 옵션
기본 사용: file filename

예: file test.txt

 test.txt: ASCII text

파일 유형만 표시 (-b): 파일명 없이 형식만 출력하여 스크립트 작성 시 유리합니다.

예: file -b test.jpg

MIME 타입 표시 (-i): text/plain, image/jpeg 등 표준 MIME 문자열로 출력합니다.

예: file -i test.sh

 test.sh: text/x-shellscript; charset=utf-8

여러 파일 확인 (-f): 목록이 담긴 파일을 읽어 일괄 확인합니다.

예: file -f file_list.txt

이 명령어는 파일 내부의 매직파일(/usr/share/magic) 데이터를 기반으로 파일을 판별하여 정확도가 높습니다.

# history 명령어
리눅스에서 history 명령어는 사용자가 이전에 실행했던 명령어 목록을 확인하고 재실행할 때 사용합니다.

터미널에 history를 입력하여 내역을 조회하며,

!번호로 특정 명령을 재실행하거나 !문자열로 가장 최근 해당 문자열로 시작한 명령어를 다시 실행할 수 있습니다.

# 주요 활용법
명령어 조회: history 명령어를 통해 실행했던 명령어들의 내역을 확인할 수 있습니다.

최근 명령어 확인: history | tail -n 10 명령어로 가장 최근 실행한 10개의 명령어를 확인할 수 있습니다.

특정 명령어 재실행: !행번호 (예: !427)를 사용하여 해당 번호의 명령어를 실행할 수 있습니다.

검색 및 재실행: Ctrl + R을 누르고 검색어를 입력하여 이전에 사용한 명령어를 역순으로 검색하고 실행할 수 있습니다.

히스토리 삭제: history -c 명령어로 전체 히스토리를 지우거나,

history -d [행번호]로 특정 명령어만 삭제할 수 있습니다.

위치: 명령어 이력은 일반적으로 $HOME/.bash_history 파일에 저장됩니다.

시간 기록: export HISTTIMEFORMAT="%F %T "를 설정하면 history 명령어 출력 시 명령어 실행 시간도 확인할 수 있습니다.

# nano 명령어

nano [파일명]으로 파일을 열고, 방향키로 이동하며, 하단 메뉴(^ = Ctrl 키)를 사용해 수정/저장/종료합니다.

주요 단축키는 저장 Ctrl+O, 종료 Ctrl+X, 잘라내기 Ctrl+K, 붙여넣기 Ctrl+U입니다.

리눅스 vi(또는 vim) 편집기는 명령 모드, 입력 모드, 마지막 행 모드로 작동하며,

i로 입력 시작, Esc로 명령 모드 전환, :wq로 저장 후 종료합니다.

dd(줄 삭제), p(붙여넣기), u(실행 취소), GG(맨 끝 이동) 등 단축키를 활용하여 파일을 효율적으로 편집합니다.

# 1. vi 실행 및 파일 열기 (쉘 명령어)
vi [파일명]: 파일 열기 또는 생성

vi +[행번호] [파일명]: 해당 행으로 커서 이동 후 열기

# 2. 입력 모드 전환 (명령 모드 -> 입력 모드)
i: 커서 현재 위치부터 입력

a: 커서 다음 위치부터 입력

o: 커서 아래에 새로운 줄 생성 후 입력

Esc: 입력 모드에서 명령 모드로 복귀

# 3. 명령 모드 명령어 (기본 모드)

p: 붙여넣기

G: 파일 맨 끝으로 이동

gg: 파일 맨 처음으로 이동

$: 줄의 맨 뒤로 이동

0: 줄의 맨 앞으로 이동
# 4. 마지막 행 모드 (저장 및 종료, 콜론: 입력 후 사용)

/[문자열]: 문자열 검색 (n은 다음 찾기, N은 이전 찾기)

:%s/old/new/g: 파일 전체에서 old를 new로 치환

Ctrl + f: 한 화면 뒤로 (Forward)

Ctrl + b: 한 화면 앞으로 (Backward)

# chmod 명령어
리눅스 chmod(change mode) 명령어는 파일이나 디렉토리의 접근 권한(읽기, 쓰기, 실행)을 변경하는 데 사용됩니다.

chmod [옵션] 모드 대상 형식으로 사용하며, 크게 숫자(8진수)를 사용하는 방식과 기호(심볼릭)를 사용하는 방식으로 나뉩니다.

# 핵심 권한 및 사용법
권한: r=4 (읽기), w=2 (쓰기), x=1 (실행).

대상: u (user), g (group), o (other), a (all).

# 주요 예시:
chmod 755 file: 파일 소유자는 모든 권한, 그룹과 나머지는 읽기/실행 권한.

chmod 644 file: 소유자는 읽기/쓰기, 그룹/나머지는 읽기 전용.

chmod +x script.sh: 모든 사용자에게 실행 권한 추가.

# 1. 숫자(8진수) 모드 (Absolute Mode)
3자리 숫자를 사용하여 소유자(User), 그룹(Group), 기타 사용자(Others)의 권한을 0~7까지 조합하여 설정합니다.

chmod 777 filename: 모든 사용자에게 모든 권한 부여 (위험).

chmod 644 filename: 파일에 대해 소유자는 읽기/쓰기, 그룹과 나머지는 읽기만 가능하게 설정.

# 2. 심볼릭 모드 (Symbolic Mode)
기호를 사용하여 직관적으로 권한을 추가(+)하거나 제거(-), 또는 설정(=)합니다.
chmod u+x file: 사용자(u)에게 실행 권한(+) 추가.

chmod g-w file: 그룹(g)에서 쓰기 권한(-) 제거.

chmod o=r file: 다른 사용자(o)에게 읽기 권한(=)만 설정.

chmod a+x file: 모든 사용자(a)에게 실행 권한(+) 추가.
# 3. 디렉토리 권한 설정
디렉토리에 실행(x) 권한이 있어야 내부 파일에 접근(cd)할 수 있습니다.

하위 디렉토리까지 모두 적용하려면 -R 옵션을 사용합니다.

chmod -R 755 directory/

# chown 명령어
리눅스 chown (change owner) 명령어는 파일이나 디렉토리의 소유자(User) 및 그룹(Group)을 변경하는 데 사용됩니다.

관리자 권한(sudo)이 필요하며, 시스템 보안 및 접근 권한 관리에 필수적인 명령어입니다.
# 핵심 사용법:
소유자만 변경: chown 사용자명 파일명

소유자와 그룹 동시 변경: chown 사용자명:그룹명 파일명

하위 디렉토리 포함 모두 변경: chown -R 사용자명:그룹명 디렉토리명 (-R은 재귀적(Recursive) 변경)

# 주요 예시:
sudo chown root:root file.txt: 파일 소유자와 그룹을 root로 변경

sudo chown -R www-data:www-data /var/www/html: 디렉토리의 모든 파일 소유권을 www-data로 변경

# ln 명령어
리눅스 ln 명령어는 파일이나 디렉토리에 대한 링크(바로가기)를 생성하는 명령어입니다.

크게 심볼릭 링크(ln -s, 바로가기)와 하드 링크(ln, 복사본)로 나뉘며,

원본 위치를 바꾸지 않고 다른 경로에서 접근하거나 데이터를 공유할 때 사용합니다.

기본 구문: ln [옵션] [원본파일] [링크파일]
# 1. 심볼릭 링크 (Symbolic Link / Soft Link) - 주로 사용
특징: 원본 파일을 가리키는 포인터(바로가기). 원본 삭제 시 링크는 깨짐. 디렉토리 링크 가능.

명령어: ln -s [원본] [링크명]

예시: ln -s /path/to/original /home/user/shortcut

# 2. 하드 링크 (Hard Link)
특징: 원본과 동일한 inode를 공유하는 실제 파일. 원본 삭제해도 링크 데이터 유지. 디렉토리 링크 불가.

명령어: ln [원본] [링크명]

예시: ln original.txt hardlink.txt

# 3. 주요 옵션
-s: 심볼릭 링크(Symbolic link) 생성

-f: 기존에 동일한 이름의 링크 파일이 존재해도 강제로 덮어쓰기

-v: 링크 생성 과정 출력

# 4. 링크 삭제
rm 명령어를 사용하여 일반 파일처럼 삭제하면 됩니다. (심볼릭 링크 삭제 시 원본은 안전)

# apt update 명령어
리눅스 sudo apt update 명령어는

저장소의 최신 패키지 목록을 가져와서 로컬 시스템의 패키지 인덱스를 갱신합니다.

실제 소프트웨어 업그레이드는 수행하지 않으며,

설치된 패키지 정보를 최신 상태로 만들어 upgrade 명령어를 위한 준비 작업을 합니다.

# 핵심 내용 및 사용법
용도: 패키지 저장소 목록 업데이트 (패키지 정보 갱신)

실제 업그레이드 여부: 아니요 (파일 다운로드/설치 안 함)

명령어: sudo apt update (관리자 권한 필요)

# 많이 함께 쓰는 명령어 (권장)
sudo apt update && sudo apt upgrade -y: 패키지 목록 업데이트 후 최신 버전을 설치

# 주의사항
apt update만으로는 설치된 프로그램이 최신 버전으로 바뀌지 않습니다.

패키지를 실제로 업데이트하려면 sudo apt upgrade를 실행해야 합니다.


# apt upgrade 명령어
리눅스에서 sudo apt update && sudo apt upgrade -y 명령어를 통해 설치된 패키지들의

최신 버전 업데이트를 안전하게 수행할 수 있습니다.

update로 최신 패키지 목록을 가져온 후,

upgrade로 실제 패키지를 업데이트하여 시스템 보안과 안정성을 유지합니다.
# 핵심 정리
명령어: sudo apt update (목록 갱신) 후 sudo apt upgrade (설치된 패키지 업그레이드).

통합 실행: sudo apt update && sudo apt upgrade -y (업데이트 목록 확인 후 바로 설치).

주요 차이: apt upgrade는 패키지를 최신 버전으로 업그레이드하지만,

필수 패키지를 제거하거나 새로운 패키지를 설치하여 의존성 문제를 해결하지는 않습니다.

주의사항: 패키지 설치 중 대화형 프롬프트(Y/N)를 생략하려면 -y 옵션을 사용합니다.

추가 명령어: 전체 시스템 업데이트를 위해 sudo apt full-upgrade를 사용할 수도 있으나,

기존 패키지 제거/설치가 포함되므로 주의가 필요합니다.


# apt install 명령어

 터미널에서 sudo apt update로 목록을 갱신한 후 sudo apt install [패키지명]을 입력합니다.

의존성 문제를 자동으로 해결하며, 설치 시 관리자 권한(sudo)이 필요합니다.

# 주요 apt 설치 및 관리 명령어:
패키지 목록 업데이트: sudo apt update (설치 전 필수)

패키지 설치: sudo apt install [패키지명]

패키지 설치 (질문 없이): sudo apt -y install [패키지명]

패키지 업그레이드: sudo apt upgrade

패키지 삭제: sudo apt remove [패키지명] (설정 파일 유지)

패키지 완전 삭제: sudo apt purge [패키지명] (설정 파일까지 제거)

설치된 패키지 목록: apt list --installed

주요 사용법 및 상세 설명
기본 명령어: tar cvf [생성할 파일명.tar] [압축 대상]

c (create): 새로운 tar 아카이브 파일을 생성
v (verbose): 묶이는 파일 목록을 화면에 표시 (과정 확인)
f (file): 아카이브 파일의 이름을 지정 (이 옵션 바로 뒤에 파일명 기술)

참고 사항

파일 묶기 (tar): tar cvf target.tar target/
압축 풀기 (tar): tar xvf target.tar


리눅스에서 tar xvf 명령어는 .tar 아카이브 파일의 압축을 풀고 내용을 화면에 표시(verbose)하며 현재 디렉토리에 해제합니다. x(eXtract, 압축 해제), v(Verbose, 진행 상황 출력), f(File, 파일명 지정) 옵션을 조합하여 주로 사용합니다.

주요 사용법 및 예시:
기본 압축 해제: tar xvf 파일명.tar
지정 디렉토리에 해제: tar xvf 파일명.tar -C /path/to/folder
.tar.gz 파일 해제: tar zxvf 파일명.tar.gz (z 옵션 추가)
.tar.bz2 파일 해제: tar jxvf 파일명.tar.bz2 (j 옵션 추가)

옵션 상세 설명:
-x (extract): 묶인 파일을 풉니다.
-v (verbose): 파일 처리 과정을 화면에 자세히 보여줍니다.
-f (file): 아카이브 파일명을 지정합니다. 반드시 가장 마지막에 위치해야 합니다.


# find 명령어
리눅스의 find 명령어는 파일 시스템에서 이름, 크기, 시간, 권한 등 다양한 조건으로

파일이나 디렉터리를 검색하는 강력한 도구입니다.

기본 구조는 find [경로] [옵션] [표현식]이며,

대표적으로 -name(이름), -type(파일/디렉터리), -size(크기), -exec(명령어 실행) 옵션을 주로 사용합니다.
# 주요 find 명령어 사용 예시
현재 디렉터리에서 파일명으로 찾기: find . -name "test.txt"

현재 디렉터리에서 대소문자 구분 없이 찾기: find . -iname "test.txt"

파일 유형으로 찾기: find . -type d (디렉터리만), find . -type f (일반 파일만)

특정 확장자 파일 찾기: find /home -name "*.log"

크기가 100MB 이상인 파일 찾기: find / -size +100M

수정된 지 7일 이상 된 파일 삭제: find /var/log -type f -mtime +7 -delete

파일 찾아서 상세 정보 출력: find . -name "*.conf" -ls

# 핵심 옵션 정리
-name / -iname: 파일 이름(pattern)으로 검색 / 대소문자 무시

-type: f(일반 파일), d(디렉터리), l(심볼릭 링크)

-size: +n (n보다 큼), -n (n보다 작음), n (정확히 n)

-mtime / -atime / -ctime: 수정/접근/변경 시간 기준 (+n: n일 이전, -n: n일 이내)

-exec: 찾은 파일에 다른 명령어를 실행 (find . -name "*.sh" -exec chmod +x {} \;)

-user / -group: 소유자/소유 그룹 기준 검색

-maxdepth: 검색할 하위 디렉터리 깊이 제한
