# ls 명령어 설명
ls 명령어는 유닉스 계열 운영 체제(Linux 등)에서 현재 디렉토리 또는 지정한 디렉토리 내의 파일과 디렉토리 목록을 표시하는 필수 명령어입니다.

list segments의 약자로, 파일명, 디렉토리 구조, 그리고 -l 옵션을 통한 권한, 소유자, 크기 등 상세 정보를 확인하는 데 사용됩니다.

# 자주 사용하는 ls 옵션
ls -l: 파일의 상세 정보(권한, 소유자, 크기, 수정 시간)를 나열합니다.

ls -a: 숨김 파일(점.으로 시작하는 파일)을 포함한 모든 파일을 표시합니다.

ls -al: 상세 정보와 함께 숨김 파일을 모두 보여줍니다.

ls -h: -l 옵션과 함께 사용하여 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시합니다.

ls -t: 파일을 수정 시간 순서대로 최신순으로 정렬하여 표시합니다.

ls -r: 목록을 역순으로 정렬합니다.

ls -R: 하위 디렉토리의 내용까지 재귀적으로 모두 나열합니다.

ls -d: 디렉토리 자체의 정보를 보여줍니다.

# 활용 예시
ls -alF: 모든 파일을 상세 정보와 함께 표시하고, 파일 형식을 알리는 기호(*, / 등)를 끝에 붙입니다.

# CD 명령어
리눅스의 cd(Change Directory) 명령어는 현재 작업 중인 디렉토리(폴더)를 이동할 때 사용하는 가장 기초적인 명령어입니다.

cd ..으로 상위 폴더로, cd 또는 cd ~으로 홈 디렉토리로 바로 이동할 수 있습니다.

cd /: 루트(/) 디렉토리로 이동

cd ~ 또는 cd: 현재 사용자의 홈 디렉토리로 이동

cd -: 바로 이전에 위치했던 디렉토리로 이동

현재 위치는 pwd 명령어로 확인 가능합니다.

# touch 명령어
리눅스의 touch 명령어는 0바이트 빈 파일을 생성하거나,

기존 파일의 최근 접근/수정 시간(atime, mtime)을 현재 시간으로 갱신하는 데 사용됩니다.

파일이 없으면 새로 만들고, 있으면 시간만 변경하며 touch [파일명] 형태로 사용합니다.

# 주요 기능 및 사용법
빈 파일 생성: touch newfile.txt (파일이 없으면 생성)

시간 갱신: touch existingfile.txt (이미 존재하는 파일의 시간 정보를 현재로 업데이트)

여러 파일 생성: touch file1.txt file2.txt (공백으로 구분하여 동시에 여러 파일 생성)

# 주요 옵션
-c: 파일이 없어도 새로 생성하지 않습니다. (시간 업데이트만 수행)

-a: 최근 접근 시간(access time)만 변경합니다.

-m: 최근 변경 시간(modification time)만 변경합니다.

-t: [[CC]YY]MMDDhhmm[.ss] 형식으로 시간을 직접 지정하여 변경합니다.

# rm 명령어
리눅스의 rm (remove) 명령어는 파일이나 디렉토리를 삭제하는 핵심 명령어입니다.

기본적으로 파일을 삭제하며, -r (디렉토리 삭제), -f (강제 삭제) 등의 옵션을 사용하여

하위 디렉토리를 포함한 폴더 전체를 강제로 지울 수 있습니다.

삭제된 파일은 복구가 어렵기 때문에 주의가 필요합니다.
# 주요 사용법 및 옵션
파일 삭제: rm [파일명] (기본적으로 삭제 여부 확인)

-i (Interactively): 삭제 전 물어봄 (rm -i file.txt)

-r (Recursive): 디렉토리 및 내부 파일 전체 삭제 (rm -r folder)

-f (Force): 존재하지 않는 파일은 무시하고, 묻지 않고 강제 삭제 (rm -f file.txt)

-rf: 디렉토리를 묻지 않고 강제로 삭제 (rm -rf folder)

# cp 명령어
리눅스 cp (copy) 명령어는 파일이나 디렉터리를 복사할 때 사용하며, 기본 구문은 cp [옵션] [원본] [대상]입니다.

파일 복사, 이름 변경 복사, 디렉터리 복사(-r), 덮어쓰기 확인(-i), 권한 유지(-p) 등

다양한 옵션을 지원하여 시스템 관리에서 필수적인 명령어입니다.
# 주요 cp 명령어 사용법 및 예제
파일 복사: cp file1.txt file1_bak.txt (file1을 file1_bak로 복사)

파일을 디렉터리로 복사: cp file1.txt /home/user/documents/

디렉터리 복사 (하위 포함): cp -r dir1 dir2 (-r 옵션 필수)

복사 시 덮어쓰기 확인: cp -i file1.txt file2.txt (대상 파일 존재 시 물어봄)

파일 속성(권한) 유지하며 복사: cp -p file1.txt file1_bak.txt

여러 파일 복사: cp file1.txt file2.txt /dest/dir/

# 자주 사용하는 핵심 옵션
-r, -R: 디렉터리를 복사할 때 하위 디렉터리와 파일을 모두 복사

-i: 복사할 목적지에 동일한 이름의 파일이 존재할 경우, 덮어쓸지 물어봄

-f: 목적지에 파일이 있어도 강제로 덮어씀

-p: 원본 파일의 권한, 소유자, 시간 등 속성을 보존

-v: 복사 과정을 화면에 표시 (verbose)

# mv 명령어
리눅스 mv (move) 명령어는 파일이나 디렉토리를 다른 위치로 이동하거나 이름을 변경할 때 사용합니다.

기본 문법은 mv [옵션] <원본> <대상>이며, 같은 디렉토리 내에서 대상을 다르게 지정하면 이름이 변경되고,

다른 디렉토리 경로를 지정하면 파일이 이동됩니다.
# 주요 활용법 및 명령어:
파일 이름 변경: mv old_name.txt new_name.txt

파일 위치 이동: mv file.txt /path/to/destination/

디렉토리 이동: mv /source_dir /path/to/destination/

여러 파일 이동: mv file1.txt file2.txt /dest_dir/

# 주요 옵션:
-i (interactive): 덮어쓰기 전 확인 메시지를 표시

-f (force): 덮어쓰기 확인 없이 강제 이동

-n (no-clobber): 대상 파일이 존재하면 덮어쓰지 않음

-v (verbose): 이동 중인 파일 목록을 상세히 출력

# 주의사항:
동일한 이름의 파일이 목적지에 있으면 경고 없이 덮어씌워지므로 주의가 필요합니다.

-i 옵션을 사용하여 덮어쓰기 전에 확인하는 것이 안전합니다.

# mkdir 명령어
리눅스 mkdir(make directory)은 새로운 디렉토리(폴더)를 생성하는 명령어입니다.

mkdir [옵션] 디렉토리명 형식을 사용하며, -p 옵션으로 상위 디렉토리까지 한번에 생성하거나, -m으로 권한을 설정할 수 있습니다.

# 주요 사용법 및 옵션
기본 생성: mkdir folder1 (현재 위치에 folder1 생성)

여러 디렉토리 생성: mkdir dir1 dir2 dir3 (공백으로 구분)

하위 디렉토리 포함 생성 (-p): mkdir -p parent/child (상위 디렉토리 없어도 자동 생성)

권한 지정하여 생성 (-m): mkdir -m 755 new_folder (생성과 동시에 권한 설정)

생성 내용 출력 (-v): mkdir -v new_folder (디렉토리 생성 메시지 출력)

# 주의사항:
동일한 이름의 디렉토리가 있으면 에러가 발생하므로, 기존 디렉토리가 있어도 에러 없이 진행하려면 -p 옵션을 활용합니다.

# cat 명령어
리눅스의 cat(concatenate) 명령어는 파일 내용을 터미널에 출력하거나,

여러 파일을 합치고(연결), 새로운 파일을 생성하는 가장 기본적인 텍스트 처리 도구입니다.

주요 사용법은 cat [파일]로 내용을 확인하거나, > 연산자와 함께 사용하여 파일 생성/저장에 활용합니다.

# 주요 기능 및 사용 예제
파일 내용 출력: cat file.txt (file.txt의 내용을 화면에 표시)

파일 내용 번호와 함께 출력: cat -n file.txt (-n 옵션 사용)

여러 파일 합치기: cat file1.txt file2.txt > combined.txt (두 파일을 합쳐서 새로운 파일로 저장)

파일 내용 추가하기: cat file1.txt >> file2.txt (file1 내용을 file2 끝에 덧붙임)

탭/줄 끝 표시: cat -T file.txt (탭 문자), cat -E file.txt (줄 끝)

cat은 작은 크기의 파일 내용을 빠르게 확인할 때 매우 유용하며,

grep이나 more와 같은 다른 명령어와 파이프(|)로 연결하여 자주 사용합니다.

# file 명령어
리눅스 file 명령어는 파일의 확장자나 이름에 의존하지 않고,

파일의 내용(헤더/매직넘버)을 분석하여 정확한 파일 형식(ASCII 텍스트, 바이너리, 이미지 등)을 식별하는 도구입니다.

file [옵션] 파일명 형식으로 사용하며, 주요 옵션으로 -b(유형만 표시), -i(MIME 타입 표시)가 있습니다.

# 주요 활용법 및 옵션
기본 사용: file filename

예: file test.txt

 test.txt: ASCII text

예: file -i test.sh

예: file -f file_list.txt

# history 명령어

터미널에 history를 입력하여 내역을 조회하며,

!번호로 특정 명령을 재실행하거나 !문자열로 가장 최근 해당 문자열로 시작한 명령어를 다시 실행할 수 있습니다.

# 주요 활용법
명령어 조회: history 명령어를 통해 실행했던 명령어들의 내역을 확인할 수 있습니다.

최근 명령어 확인: history | tail -n 10 명령어로 가장 최근 실행한 10개의 명령어를 확인할 수 있습니다.

특정 명령어 재실행: !행번호 (예: !427)를 사용하여 해당 번호의 명령어를 실행할 수 있습니다.

검색 및 재실행: Ctrl + R을 누르고 검색어를 입력하여 이전에 사용한 명령어를 역순으로 검색하고 실행할 수 있습니다.

히스토리 삭제: history -c 명령어로 전체 히스토리를 지우거나,

history -d [행번호]로 특정 명령어만 삭제할 수 있습니다.

리눅스 vi(또는 vim) 편집기는 명령 모드, 입력 모드, 마지막 행 모드로 작동하며,

i로 입력 시작, Esc로 명령 모드 전환, :wq로 저장 후 종료합니다.

dd(줄 삭제), p(붙여넣기), u(실행 취소), GG(맨 끝 이동) 등 단축키를 활용하여 파일을 효율적으로 편집합니다.

# 1. vi 실행 및 파일 열기 (쉘 명령어)
vi [파일명]: 파일 열기 또는 생성

vi +[행번호] [파일명]: 해당 행으로 커서 이동 후 열기

# 2. 입력 모드 전환 (명령 모드 -> 입력 모드)
i: 커서 현재 위치부터 입력

a: 커서 다음 위치부터 입력

o: 커서 아래에 새로운 줄 생성 후 입력

Esc: 입력 모드에서 명령 모드로 복귀

G: 파일 맨 끝으로 이동

$: 줄의 맨 뒤로 이동

0: 줄의 맨 앞으로 이동
# 4. 마지막 행 모드 (저장 및 종료, 콜론: 입력 후 사용)

Ctrl + b: 한 화면 앞으로 (Backward)

# chmod 명령어
리눅스 chmod(change mode) 명령어는 파일이나 디렉토리의 접근 권한(읽기, 쓰기, 실행)을 변경하는 데 사용됩니다.

chmod [옵션] 모드 대상 형식으로 사용하며, 크게 숫자(8진수)를 사용하는 방식과 기호(심볼릭)를 사용하는 방식으로 나뉩니다.

# 핵심 권한 및 사용법
권한: r=4 (읽기), w=2 (쓰기), x=1 (실행).

대상: u (user), g (group), o (other), a (all).

# 주요 예시:
chmod 755 file: 파일 소유자는 모든 권한, 그룹과 나머지는 읽기/실행 권한.

chmod 644 file: 소유자는 읽기/쓰기, 그룹/나머지는 읽기 전용.

chmod +x script.sh: 모든 사용자에게 실행 권한 추가.

# 1. 숫자(8진수) 모드 (Absolute Mode)
3자리 숫자를 사용하여 소유자(User), 그룹(Group), 기타 사용자(Others)의 권한을 0~7까지 조합하여 설정합니다.

chmod 777 filename: 모든 사용자에게 모든 권한 부여 (위험).

chmod 644 filename: 파일에 대해 소유자는 읽기/쓰기, 그룹과 나머지는 읽기만 가능하게 설정.

# 2. 심볼릭 모드 (Symbolic Mode)
기호를 사용하여 직관적으로 권한을 추가(+)하거나 제거(-), 또는 설정(=)합니다.
chmod u+x file: 사용자(u)에게 실행 권한(+) 추가.

chmod g-w file: 그룹(g)에서 쓰기 권한(-) 제거.

chmod o=r file: 다른 사용자(o)에게 읽기 권한(=)만 설정.

chmod a+x file: 모든 사용자(a)에게 실행 권한(+) 추가.
# 3. 디렉토리 권한 설정
디렉토리에 실행(x) 권한이 있어야 내부 파일에 접근(cd)할 수 있습니다.

하위 디렉토리까지 모두 적용하려면 -R 옵션을 사용합니다.

chmod -R 755 directory/

# chown 명령어
리눅스 chown (change owner) 명령어는 파일이나 디렉토리의 소유자(User) 및 그룹(Group)을 변경하는 데 사용됩니다.

관리자 권한(sudo)이 필요하며, 시스템 보안 및 접근 권한 관리에 필수적인 명령어입니다.
# 핵심 사용법:
소유자만 변경: chown 사용자명 파일명

소유자와 그룹 동시 변경: chown 사용자명:그룹명 파일명

하위 디렉토리 포함 모두 변경: chown -R 사용자명:그룹명 디렉토리명 (-R은 재귀적(Recursive) 변경)

# 주요 예시:
sudo chown root:root file.txt: 파일 소유자와 그룹을 root로 변경

sudo chown -R www-data:www-data /var/www/html: 디렉토리의 모든 파일 소유권을 www-data로 변경

# ln 명령어
리눅스 ln 명령어는 파일이나 디렉토리에 대한 링크(바로가기)를 생성하는 명령어입니다.

크게 심볼릭 링크(ln -s, 바로가기)와 하드 링크(ln, 복사본)로 나뉘며,

원본 위치를 바꾸지 않고 다른 경로에서 접근하거나 데이터를 공유할 때 사용합니다.

기본 구문: ln [옵션] [원본파일] [링크파일]
# 1. 심볼릭 링크 (Symbolic Link / Soft Link) - 주로 사용
특징: 원본 파일을 가리키는 포인터(바로가기). 원본 삭제 시 링크는 깨짐. 디렉토리 링크 가능.

명령어: ln -s [원본] [링크명]

예시: ln -s /path/to/original /home/user/shortcut

# 2. 하드 링크 (Hard Link)
특징: 원본과 동일한 inode를 공유하는 실제 파일. 원본 삭제해도 링크 데이터 유지. 디렉토리 링크 불가.

명령어: ln [원본] [링크명]

예시: ln original.txt hardlink.txt

# 3. 주요 옵션
-s: 심볼릭 링크(Symbolic link) 생성

-f: 기존에 동일한 이름의 링크 파일이 존재해도 강제로 덮어쓰기

-v: 링크 생성 과정 출력

# 4. 링크 삭제
rm 명령어를 사용하여 일반 파일처럼 삭제하면 됩니다. (심볼릭 링크 삭제 시 원본은 안전)

# apt update 명령어
리눅스 sudo apt update 명령어는

저장소의 최신 패키지 목록을 가져와서 로컬 시스템의 패키지 인덱스를 갱신합니다.

실제 소프트웨어 업그레이드는 수행하지 않으며,

설치된 패키지 정보를 최신 상태로 만들어 upgrade 명령어를 위한 준비 작업을 합니다.

# 핵심 내용 및 사용법
용도: 패키지 저장소 목록 업데이트 (패키지 정보 갱신)

실제 업그레이드 여부: 아니요 (파일 다운로드/설치 안 함)

명령어: sudo apt update (관리자 권한 필요)

# 많이 함께 쓰는 명령어 (권장)
sudo apt update && sudo apt upgrade -y: 패키지 목록 업데이트 후 최신 버전을 설치

# 주의사항
apt update만으로는 설치된 프로그램이 최신 버전으로 바뀌지 않습니다.

패키지를 실제로 업데이트하려면 sudo apt upgrade를 실행해야 합니다.


# apt upgrade 명령어
리눅스에서 sudo apt update && sudo apt upgrade -y 명령어를 통해 설치된 패키지들의

최신 버전 업데이트를 안전하게 수행할 수 있습니다.

update로 최신 패키지 목록을 가져온 후,

upgrade로 실제 패키지를 업데이트하여 시스템 보안과 안정성을 유지합니다.
# 핵심 정리
명령어: sudo apt update (목록 갱신) 후 sudo apt upgrade (설치된 패키지 업그레이드).

통합 실행: sudo apt update && sudo apt upgrade -y (업데이트 목록 확인 후 바로 설치).

주요 차이: apt upgrade는 패키지를 최신 버전으로 업그레이드하지만,

필수 패키지를 제거하거나 새로운 패키지를 설치하여 의존성 문제를 해결하지는 않습니다.

주의사항: 패키지 설치 중 대화형 프롬프트(Y/N)를 생략하려면 -y 옵션을 사용합니다.

추가 명령어: 전체 시스템 업데이트를 위해 sudo apt full-upgrade를 사용할 수도 있으나,

기존 패키지 제거/설치가 포함되므로 주의가 필요합니다.

# find 명령어
리눅스의 find 명령어는 파일 시스템에서 이름, 크기, 시간, 권한 등 다양한 조건으로

파일이나 디렉터리를 검색하는 강력한 도구입니다.

기본 구조는 find [경로] [옵션] [표현식]이며,

대표적으로 -name(이름), -type(파일/디렉터리), -size(크기), -exec(명령어 실행) 옵션을 주로 사용합니다.
# 주요 find 명령어 사용 예시
현재 디렉터리에서 파일명으로 찾기: find . -name "test.txt"

현재 디렉터리에서 대소문자 구분 없이 찾기: find . -iname "test.txt"

파일 유형으로 찾기: find . -type d (디렉터리만), find . -type f (일반 파일만)

특정 확장자 파일 찾기: find /home -name "*.log"

크기가 100MB 이상인 파일 찾기: find / -size +100M

수정된 지 7일 이상 된 파일 삭제: find /var/log -type f -mtime +7 -delete

파일 찾아서 상세 정보 출력: find . -name "*.conf" -ls

# 핵심 옵션 정리
-name / -iname: 파일 이름(pattern)으로 검색 / 대소문자 무시

-type: f(일반 파일), d(디렉터리), l(심볼릭 링크)

-size: +n (n보다 큼), -n (n보다 작음), n (정확히 n)

-mtime / -atime / -ctime: 수정/접근/변경 시간 기준 (+n: n일 이전, -n: n일 이내)

-exec: 찾은 파일에 다른 명령어를 실행 (find . -name "*.sh" -exec chmod +x {} \;)

-user / -group: 소유자/소유 그룹 기준 검색

-maxdepth: 검색할 하위 디렉터리 깊이 제한

# systemctl 명령어
리눅스 systemctl은 systemd 시스템 및 서비스 관리자(PID 1)를 제어하는 핵심 명령어로,

서비스의 시작, 중지, 다시 시작, 상태 확인, 부팅 시 자동 실행 설정 등을 담당합니다.

서비스, 마운트, 소켓 등 유닛(Unit) 단위로 시스템 리소스를 관리하며

/etc/systemd/system/에 정의된 파일들을 통해 작동합니다.

# 핵심 명령어 및 활용법:
서비스 상태 확인: systemctl status [서비스명] (예: systemctl status nginx)

서비스 시작/중지: systemctl start [서비스명], systemctl stop [서비스명]

서비스 다시 시작: systemctl restart [서비스명]

부팅 시 자동 실행 설정/해제: systemctl enable [서비스명], systemctl disable [서비스명]

실행 중인 서비스 목록: systemctl list-units --type=service --state=running

서비스 재로드(설정 반영): systemctl reload [서비스명]

systemctl은 이전의 service 명령어를 대체하여 더욱 빠르고 효율적인 병렬 서비스 시작을 지원합니다.


# ifconfig 명령어
리눅스 ifconfig는 네트워크 인터페이스(NIC) 정보를

확인, 설정, 활성화/비활성화(up/down)하는 기본 명령어로,

IP 주소, MAC 주소, 서브넷 마스크, 브로드캐스트 주소 등을 표시합니다.

현재는 ip addr` 명령어로 대체되는 추세입니다.
# 핵심 사용법
전체 정보 확인: ifconfig
특정 인터페이스 확인: ifconfig ens32 or 33 (예: eth0)

인터페이스 비활성화: ifconfig ens32 down

인터페이스 활성화: ifconfig ens32 up

IP 및 넷마스크 설정: ifconfig ens32 192.168.1.10 netmask 255.255.255.0
# 주의사항
대부분의 작업은 sudo (관리자 권한)가 필요합니다.

최신 리눅스 배포판에서는 net-tools 패키지 미설치로 ifconfig 명령어가 없을 수 있으며,

이 경우 sudo apt install net-tools로 설치하거나 ip 명령어를 사용해야 합니다.

# ping 명령어
리눅스 ping은 ICMP 프로토콜을 사용하여 원격 호스트와의

네트워크 연결 상태(연결 가능 여부, 지연 시간(latency), 패킷 손실률)를 확인하는

필수 진단 명령어입니다.

기본적으로 무한 전송되므로 Ctrl+C로 중단하며, -c 옵션으로 횟수를 제한할 수 있습니다.

# 핵심 사용법 및 주요 옵션
기본 사용법: ping [호스트명 또는 IP] (예: ping google.com, ping 8.8.8.8)

횟수 제한: ping -c [횟수] [호스트] (예: ping -c 4 google.com) - 4번만 핑을 보내고 멈춤.

간격 조절: ping -i [초] [호스트] (예: ping -i 0.5 google.com) - 패킷 전송 간격을 초 단위로 조절.

핑 종료: Ctrl+C 키 조합을 눌러 핑 전송을 멈추고 통계 확인.

# 출력 결과 해석
64 bytes from ...: 응답을 성공적으로 받았음을 의미.

time=XXms: 응답 속도(레이턴시)로, 숫자가 작을수록 연결이 빠름.

packet loss: 손실된 패킷 비율. 0%가 이상적이며, 높으면 네트워크 불안정.

# 주요 활용 팁
연결 확인: ping 168.126.63.1 (KT DNS) 등으로 인터넷 연결 여부를 1차적으로 진단.

도메인 확인: ping 도메인명을 통해 도메인이 올바른 IP로 매핑되는지 확인 가능.

주의사항: 방화벽(Firewall)이 ICMP를 차단하면 핑이 응답하지 않을 수 있습니다.


# nslookup 명령어
리눅스 nslookup은 DNS 서버에 질의하여

도메인의 IP 주소나 레코드 정보를 조회하는 네트워크 디버깅 도구입니다.

nslookup [도메인]으로 사용하며,

도메인의 IP 확인(A 레코드), 네임서버 조회(NS 레코드), 메일 교환 서버(MX 레코드) 확인 등에 활용됩니다.
# 주요 사용법 및 팁
기본 도메인 조회: nslookup example.com

특정 DNS 서버 지정 조회: nslookup example.com 8.8.8.8 (구글 DNS 서버를 통해 조회)

레코드 타입 지정: nslookup -type=mx example.com (MX 레코드 조회)

대화형 모드: nslookup 입력 후 프롬프트(>)에서 도메인 및 명령 입력
# 설치 방법
CentOS/RHEL/Fedora: sudo yum install bind-utils
Ubuntu/Debian: sudo apt-get install dnsutils

# grep 명령어
리눅스 grep은 파일 내에서 특정 문자열이나 정규 표현식 패턴을 검색하여

해당 줄을 출력하는 강력한 텍스트 검색 도구입니다.

grep [옵션] "패턴" [파일명] 형식으로 사용하며,

로그 분석, 파일 내 변수 찾기 등 시스템 관리 및 개발에서 핵심적인 역할을 합니다.

# 주요 옵션 및 사용법
-i: 대소문자 구분 없이 검색 (grep -i "error" log.txt)

-r (또는 -R): 하위 디렉토리를 포함한 모든 파일에서 검색 (grep -r "pattern" .)

-n: 검색 결과의 라인 번호 출력 (grep -n "main" code.c)

-v: 패턴을 포함하지 않는 라인만 출력 (grep -v "info" log.txt)

-l: 패턴이 포함된 파일명만 출력 (grep -l "error" *.log)

-E: 확장 정규 표현식(Extended Regexp) 사용 (grep -E "pattern1|pattern2")
# 실전 활용 예시
파이프(|)와 조합: ps -ef | grep "nginx" (nginx 관련 프로세스 확인)

재귀적 검색: grep -rn "TODO" . (현재 디렉토리 및 하위 폴더 전체에서 TODO 검색)

특정 파일 형식에서 검색: grep "function" *.py (python 파일에서만 검색)

기본적인 문자열 매칭 외에 정규 표현식을 활용하면

매우 복잡한 패턴도 효율적으로 찾을 수 있어 리눅스 환경에서 필수적인 명령어입니다.


# 파이프( | ) 명령어
리눅스 파이프(|)는 앞선 명령어의 표준 출력(stdout)을

다음 명령어의 표준 입력(stdin)으로 직접 전달하는

강력한 프로세스 간 통신 기능입니다.

이를 통해 여러 명령어를 연결하여

데이터 필터링, 정렬, 가공 등 복잡한 작업을 효율적으로 처리할 수 있습니다.

# 주요 특징 및 활용
기호: 세로 바(|) 문자를 사용하여 명령어들을 연결.

작동 원리: 명령어1 | 명령어2 형식으로 입력하면 명령어1의 결과가 파일에 저장되지 않고 바로 명령어2로 넘어가 처리됨.

파이프라인 구성: | 기호를 연속으로 사용하여 무제한으로 명령어를 체인처럼 묶을 수 있음.

자주 사용하는 조합: grep (필터링), sort (정렬), uniq (중복 제거),

awk/sed (가공), wc (개수 세기), less/more (화면 출력 제한).

# 실무 예시
파일에서 특정 단어 검색: cat file.txt | grep "error" (file.txt에서 error가 포함된 줄만 출력)

현재 디렉토리 파일 수 세기: ls -l | wc -l (현재 폴더의 파일 개수를 세어 출력)

프로세스 검색: ps aux | grep nginx (nginx가 실행 중인지 확인)

로그 파일 10줄만 보기: cat /var/log/syslog | tail -n 10

파이프는 임시 파일을 만들지 않아 시스템 자원을 절약하고 명령어를 모듈화하여 가독성을 높여줍니다.


# 리다이렉션 명령어
리눅스 리다이렉션(>, >>, <, 2>)은

명령의 표준 입출력(stdin, stdout, stderr)을 화면 대신 파일이나 다른 장치로 바꾸는 기능입니다.

>는 파일 덮어쓰기, >>는 파일 덧붙이기, <는 파일에서 입력받기, 2>는 에러 메시지를 파일로 저장할 때 사용하며,

작업 자동화와 결과 저장에 필수적입니다.

# 주요 리다이렉션 연산자 설명
> (표준 출력 덮어쓰기): 명령의 표준 출력을 파일로 보냅니다. 파일이 있으면 덮어쓰고, 없으면 생성합니다
.
예: ls -l > file_list.txt (ls 결과를 file_list.txt에 저장)

>> (표준 출력 덧붙이기): 명령의 표준 출력을 파일의 기존 내용 끝에 추가합니다.

예: echo "Hello" >> log.txt (log.txt 파일에 "Hello" 내용 추가)

< (표준 입력): 키보드 대신 파일의 내용을 명령의 입력으로 전달합니다.

예: sort < input.txt (input.txt의 내용을 정렬)

2> (표준 에러 리다이렉션): 명령 실행 중 발생한 오류 메시지만 파일로 보냅니다.

예: command 2> error.log (에러 발생 시 error.log에 기록)

&> (표준 출력 및 에러 모두 리다이렉션): 정상 결과와 에러 메시지를 모두 하나의 파일로 저장합니다.

예: command &> all_output.txt

/dev/null (쓰레기통): 불필요한 출력이나 에러를 없애버릴 때 사용합니다.

예: command > /dev/null 2>&1 (결과와 에러를 모두 무시)

# 참고 사항
파일 디스크립터(File Descriptor):

리눅스는 표준 입력(0, stdin), 표준 출력(1, stdout), 표준 에러(2, stderr)를 번호로 관리합니다.

파이프(|): 리다이렉션이 파일로 결과를 보낸다면, 파이프는 한 명령의 출력을 다른 명령의 입력으로 직접 연결합니다.


# ps - ef 명령어
리눅스의 ps -ef 명령어는 시스템에서 실행 중인 모든 프로세스(-e)를

상세한 형식(-f, full format)으로 화면에 출력하는 명령어입니다.

UID, PID, PPID, 실행 시간 등 전체 프로세스 상태를 파악할 때 유용하며, grep과

함께 사용하여 특정 프로세스만 검색하는 데 주로 사용됩니다.
# 주요 특징 및 상세 설명
-e (Every): 시스템의 모든 프로세스를 표시합니다.

-f (Full): 프로세스의 상세한 정보(UID, PID, PPID, CPU 사용량, 실행 시작 시간,
터미널 정보, CPU 시간, 명령어)를 전체 형식으로 표시합니다.

# 출력 항목:
UID: 프로세스를 실행한 사용자 ID.

PID: 프로세스 ID (Process ID).

PPID: 부모 프로세스 ID (Parent Process ID).

C: 프로세스 실행에 사용되는 CPU 비율.

STIME: 프로세스 시작 시간.

TTY: 프로세스가 연결된 터미널.

TIME: 프로세스가 사용한 총 CPU 시간.

CMD: 프로세스를 실행한 명령어.

# 주요 활용 예시
전체 프로세스 상세 보기: ps -ef

특정 프로세스 찾기 (예: nginx): ps -ef | grep nginx

내용이 많을 때 페이징해서 보기: ps -ef | more


# kill 명령어
리눅스 kill 명령어는 실행 중인 프로세스에 특정 시그널(Signal)을 보내

프로세스를 종료하거나 제어하는 명령어입니다.

기본적으로 프로세스 ID(PID)를 사용하여 SIGTERM(15) 시그널을 전달해

정상 종료를 유도하며, kill -9로 즉시 강제 종료할 수도 있습니다.

# 주요 내용:
기본 명령어: kill [PID] (프로세스 ID 입력, 기본값은 -15)

# 시그널 종류:

-15 (SIGTERM): 기본값. 프로세스가 정리 작업을 마치고 안전하게 종료되도록 요청합니다.

-9 (SIGKILL): 즉시 강제 종료. 프로세스가 정리 작업 없이 즉시 종료됩니다.

-1 (SIGHUP): 프로세스 재시작(설정 재로드).

# 사용 예시:
kill 1234: PID 1234 프로세스를 정상 종료

kill -9 1234: PID 1234 프로세스를 강제 종료

kill -l: 사용 가능한 시그널 목록 확인

주의사항: 자신이 소유하지 않은 프로세스를 종료하려면 root 권한(sudo)이 필요합니다.

대안: pkill (이름으로 종료), killall (프로세스 이름의 모든 인스턴스 종료)

보통 프로세스 종료 시 kill -15를 먼저 시도하여 정상적인 정리를 유도하고, 응답이 없을 때 kill -9를 사용하는 것이 좋습니다.

# fdisk 명령어
리눅스 fdisk는 디스크 파티션을 관리(생성, 삭제, 조회)하는 핵심 명령어로,

주로 하드 디스크나 SSD의 파티션 테이블을 설정할 때 사용됩니다.

sudo fdisk -l로 디스크를 확인하고, sudo fdisk /dev/sdb 형태로 실행하여

n(생성), d(삭제), p(조회), w(저장) 명령으로 파티션을 설정합니다.
# 주요 기능 및 사용법
디스크 파티션 목록 확인: sudo fdisk -l 명령을 통해 시스템의 모든 디스크와 파티션 정보를 조회합니다.

파티션 편집 모드 진입: sudo fdisk /dev/sdb (편집할 디스크 지정).
# 주요 내부 명령어:
m: 도움말 메뉴 표시.

p: 현재 파티션 테이블 상태 출력.

n: 새로운 파티션 생성 (New).

d: 파티션 삭제 (Delete).

t: 파티션 유형 변경 (Type, 예: Linux, Swap 등).

w: 파티션 테이블 저장 후 종료 (Write).

q: 저장하지 않고 종료 (Quit).

# 파티션 작업 순서 (예시)
확인: sudo fdisk -l로 디스크명(sdb, sdc 등) 확인.

편집: sudo fdisk /dev/sdb로 도구 진입.

생성: n 입력 후 엔터(기본값 사용)로 파티션 생성
.
저장: w 입력하여 변경 사항 저장.

포맷: mkfs.ext4 /dev/sdb1 등으로 파티션을 포맷.

마운트: mount /dev/sdb1 /mnt 등으로 디렉토리에 연결.

주의: fdisk는 시스템 데이터를 변경하므로 sudo 권한이 필요하며,

잘못된 파티션 삭제 시 데이터 손실이 발생할 수 있으므로 주의해야 합니다.

# mkfs.ext4 명령어
리눅스에서 mkfs.ext4 명령어는 블록 장치(파티션, LVM 등)를 ext4 파일 시스템으로 포맷하는 데 사용됩니다.

sudo mkfs.ext4 /dev/sdb1 형태로 사용하며,

대상 파티션의 데이터를 모두 삭제하고 새로운 파일 시스템을 생성하므로 주의가 필요합니다.

# 주요 내용:
기본 명령어: sudo mkfs.ext4 [장치명] (예: sudo mkfs.ext4 /dev/sdb1)

파일 시스템 유형 지정: sudo mkfs -t ext4 [장치명]

UUID 지정: sudo mkfs.ext4 -U [UUID] [장치명]

필수 조건: 명령 실행 전 파티션은 언마운트(umount) 상태여야 합니다.

# 주의사항:
/dev/sdb1과 같은 대상 장치명을 정확히 확인해야 합니다. 잘못 지정 시 기존 데이터가 모두 유실됩니다.

# df -h 명령어
df -h 명령어는 리눅스에서 마운트된 파일 시스템의

전체 용량, 사용량, 남은 공간을 GB, MB 등 보기 편한 단위(Human-readable)로

보여주는 디스크 공간 확인 명령어입니다.

디스크 파티션별 사용 현황을 신속히 파악할 수 있는 가장 대표적인 도구입니다.
# 핵심 정보
용도: 리눅스 서버의 전체 파티션(디스크) 공간 현황 확인

h 옵션: 용량을 사람이 보기 좋게 K, M, G 단위로 변환하여 표시 (1024 단위)

결과 필드:

Filesystem: 파일 시스템 장치명

Size: 전체 용량

Used: 사용 중인 용량

Avail: 남은 용량

Use%: 사용률(%)

Mounted on: 마운트된 디렉토리 위치 (루트 / 등)

# 주요 활용 예시
df -h: 기본 사용 (대부분의 환경에서 가장 선호됨)

df -hT: 파일 시스템 종류(Type)까지 함께 출력하여 확인

df -h .: 현재 디렉토리가 포함된 파티션의 용량만 확인

du -sh * 명령어와 함께 사용하면 특정 디렉토리의 용량을 확인할 때 매우 유용합니다.

# mount 명령어
리눅스 mount 명령어는 하드디스크 파티션, USB, CD/DVD, ISO 파일 등 물리적/논리적 장치를

특정 디렉토리(마운트 포인트)에 연결하여 파일 시스템으로 접근 가능하게 만드는 명령어입니다.

기본 구조는 mount [옵션] <장치명> <마운트포인트>이며, umount로 마운트를 해제합니다.
# 주요 mount 명령어 및 옵션
기본 마운트: mount /dev/sdb1 /mnt/usb (장치를 디렉토리에 연결)

파일 시스템 유형 지정: mount -t ext4 /dev/sdb1 /mnt/data (예: ext4, vfat, nfs)

읽기 전용 마운트: mount -o ro /dev/sdb1 /mnt/usb

ISO 파일 마운트: mount -o loop image.iso /mnt/iso

마운트 정보 확인: mount 또는 df -h

마운트 해제: umount /mnt/usb 또는 umount /dev/sdb1

# 핵심 팁 및 주의사항
마운트 포인트: 장치가 연결될 디렉토리는 미리 생성되어 있어야 합니다 (mkdir 사용).

영구 마운트: 부팅 시 자동으로 마운트하려면 /etc/fstab 파일에 설정을 추가해야 합니다.

언마운트 시 주의: 마운트된 디렉토리 내에 위치한 상태에서는 umount가 작동하지 않습니다.

연결 확인: lsblk 또는 fdisk -l 명령어로 장치명 확인이 가능합니다.

# 자주 사용하는 옵션
loop: 루프 장치를 사용하여 파일(ISO 등)을 마운트

remount: 이미 마운트된 파일 시스템의 옵션을 다시 마운트(변경)

bind: 디렉토리의 일부를 다른 위치로 마운트

# umount 명령어
리눅스 umount 명령어는 마운트된 파일 시스템(저장 장치)을

디렉토리 트리에서 분리(언마운트)하여 안전하게 제거하는 데 사용됩니다.

주로 sudo umount [마운트 지점/장치명] 형식을 사용하며,

루트 권한이 필요합니다.

"device is busy" 에러 발생 시 -l(lazy) 또는 -f(force) 옵션을 사용하여 강제 해제할 수 있습니다.

# 핵심 요약 및 사용법
기본 명령: sudo umount /mnt/usb (디렉토리 지정) 또는 sudo umount /dev/sdb1 (장치명 지정)

권한: 마운트 해제는 시스템 파일에 접근하므로 보통 sudo를 앞에 붙여야 합니다.

해제 확인: lsblk 명령어를 사용하여 마운트가 해제되었는지 확인할 수 있습니다.

주의사항: 해당 디렉토리를 사용하는 프로그램이 있다면 device is busy 오류가
발생합니다.

해당 프로세스를 종료 후 다시 시도해야 합니다.

# 주요 옵션
-l (Lazy): 파일 시스템이 현재 사용 중이더라도 즉시 마운트 지점을 분리하고, 나중에 파일 시스템이 사용되지 않을 때 정리를 완료합니다.

-f (Force): NFS(네트워크 파일 시스템) 등에서 마운트 해제를 강제합니다.

-v (Verbose): 마운트 해제 과정을 자세히 표시합니다.


# "device is busy" 발생 시 대처
해당 디렉토리에 접근 중인 프로세스가 없는지 lsof 명령어로 확인 후 kill 명령어로 종료하거나, umount -l을 사용합니다.


# df -h 명령어
df 명령어는 "Disk Free"의 약어로서 현재 사용중인 파일시스템의

전체크기, 사용중인 크기, 사용가능한 크기, 사용율, 마운트정보등을 보여주는 리눅스 내장 명령어 입니다.

# echo 명령어
리눅스 echo 명령어는 터미널에 문자열이나 변수 내용을 출력하는 가장 기초적인 명령어로,

쉘 스크립트에서 메시지를 표시하거나 파일에 내용을 쓸 때 주로 사용됩니다.

기본 구문은 echo [문자열]이며,

-e 옵션으로 줄바꿈(\n)이나 탭(\t) 같은 특수문자를 해석할 수 있습니다.

# 주요 활용법 및 옵션
기본 출력: echo "Hello World" (문자열 출력)

환경 변수 출력: echo $PATH (PATH 환경 변수 내용 출력)

줄바꿈 없이 출력 (-n): echo -n "Hello" (개행 문자 없이 출력하여 다음 명령어가 줄 이어서 출력되게 함)

이스케이프 문자 해석 (-e): echo -e "Line1\nLine2" (줄바꿈 해석)

파일 저장/덮어쓰기: echo "text" > file.txt (file.txt 내용을 text로 교체)

파일에 내용 추가: echo "text" >> file.txt (file.txt 끝에 text 추가)


# 따옴표 사용 팁
" (쌍따옴표): 변수($VAR)를 해석하여 출력

' (홑따옴표): 변수를 해석하지 않고 입력한 그대로 출력

# export 명령어
리눅스 export 명령어는 쉘 변수를 환경 변수로 설정하여,

현재 쉘뿐만 아니라 하위 프로세스에서도 사용할 수 있게 하는 명령입니다.

주로 PATH와 같은 환경 설정이나 API 키 등을 설정할 때 사용하며,

터미널 종료 시 설정이 사라지는 임시 방식입니다.

# 주요 활용법 및 특징
변수 설정 및 내보내기: export 변수명=값 (예: export JAVA_HOME=/usr/lib/jvm)

자식 프로세스 전파: 일반 쉘 변수와 달리, export된 변수는 그로부터 시작된 모든 스크립트나 프로그램(자식 프로세스)에서 사용 가능합니다
.
확인: export (전체 조회), echo $변수명 (특정 변수 값 조회)

삭제: unset 변수명

# 영구 적용 방법
터미널을 재시작해도 유지하려면 설정 파일을 수정해야 합니다.

사용자별 영구 적용: ~/.bashrc 또는 ~/.bash_profile (Zsh의 경우 .zshrc) 파일 끝에 export ... 내용을 추가한 후 source ~/.bashrc로 적용합니다.

전체 사용자 영구 적용: /etc/profile 또는 /etc/bash.bashrc 파일을 수정합니다.


# printenv 명령어
리눅스의 printenv 명령어는 현재 시스템에 설정된 환경 변수(Environment Variables) 목록을 출력하거나

특정 환경 변수의 값을 조회하는 데 사용됩니다.

인자 없이 사용하면 모든 환경 변수를 NAME=VALUE 형식으로 출력하며,

printenv 변수명 형태로 특정 변수의 값만 선택적으로 출력할 수 있습니다.
# 주요 사용법 및 특징
모든 환경 변수 조회: $ printenv

특정 환경 변수 조회: $ printenv [변수명] (예: $ printenv PATH)

결과 필터링 (grep 사용): $ printenv | grep [검색어] (예: $ printenv | grep USER)

긴 목록 페이징 처리: $ printenv | less

특징: 환경 변수를 출력하는 env 명령어와 유사하지만, printenv는 오직 환경 변수를 조회하는 목적에 특화되어 있습니다.

종료 코드: 지정한 변수가 존재하지 않으면 1을, 존재하면 0을 반환합니다.
# env 및 set 명령어와의 차이점
printenv/env: 환경 변수(시스템 전체 설정)를 보여줍니다.

set: 환경 변수뿐만 아니라 쉘 변수(현재 쉘에서만 유효)까지 모두 보여줍니다.


# talnet 명령어
리눅스 텔넷(Telnet)은 원격 호스트에 접속하거나 특정 포트의 오픈 여부(방화벽 확인)를 진단하는 네트워크 도구입니다.

기본 명령어는 telnet [IP 또는 도메인] [포트]이며,

설치가 안 된 경우 yum 또는 apt를 통해 설치하여 사용합니다.
# 주요 명령어 및 사용법
기본 접속: telnet 192.168.0.1 80 (IP 192.168.0.1의 80번 포트로 연결).

포트 오픈 확인: Connected... 메시지가 나오면 성공, Connection refused 또는 Timeout은 포트가 닫혀있거나 차단된 상태입니다.

설치 (Debian/Ubuntu): sudo apt install telnet.

설치 (RHEL/CentOS): sudo yum install telnet.

종료: Ctrl + ] 입력 후 quit 입력.
# 참고 사항
텔넷은 데이터를 암호화하지 않아 보안에 취약하므로,

원격 접속 용도로는 SSH를 권장하며 포트 오픈 확인용으로만 사용하는 것이 좋습니다.


# apt install openssh-server 명령어
리눅스(Ubuntu/Debian 계열)에서 원격 접속을 위한

SSH 서버 설치 명령어는 sudo apt install openssh-server입니다.

패키지 목록을 최신화한 후 설치하는 것이 권장되며,

설치 후 자동으로 서비스가 시작되거나 sudo systemctl enable --now ssh로 실행할 수 있습니다.
# 설치 및 실행 절차:
패키지 업데이트: sudo apt update

OpenSSH 설치: sudo apt install openssh-server

서비스 실행 확인: sudo systemctl status ssh

방화벽 설정(필요 시): sudo ufw allow ssh

설치가 완료되면 다른 기기에서 ssh username@IP주소를 통해 접속 가능합니다.

# systemctl enable ssh 명령어
리눅스에서 systemctl enable ssh (또는 sshd) 명령어는

SSH 서비스를 부팅 시 자동 실행되도록 설정합니다.

이 명령어는 서비스의 실행 파일 연결을 등록하며, 현재 즉시 서비스를 시작하지는 않습니다.

# 명령어 및 관련 필수 명령어
부팅 시 자동 실행 설정: sudo systemctl enable ssh (Ubuntu/Debian) 또는 sudo systemctl enable sshd (CentOS/RHEL)

현재 즉시 시작: sudo systemctl start ssh

서비스 상태 확인: sudo systemctl status ssh

부팅 시 자동 실행 해제: sudo systemctl disable ssh

보통 설치 후 sudo systemctl enable --now ssh 명령어를 사용하면

자동 실행 등록과 즉시 시작을 동시에 처리할 수 있습니다.

# systemctl restart ssh 명령어
리눅스에서 SSH 서비스를 재시작하려면 sudo systemctl restart ssh (Ubuntu/Debian 계열)

또는 sudo systemctl restart sshd (CentOS/RHEL 계열) 명령어를 사용합니다.

설정 파일 변경 후 적용하기 위해 필수적이며, 권한이 필요하므로 앞에 sudo를 붙여 실행하는 것이 좋습니다.

# 1. OS별 SSH 재시작 명령어
Ubuntu, Debian, Mint: sudo systemctl restart ssh

CentOS, RHEL, Fedora, Rocky Linux: sudo systemctl restart sshd

# 2. 서비스 상태 확인 (재시작 후)
정상적으로 다시 시작되었는지 확인하려면 다음 명령어를 사용합니다.
sudo systemctl status ssh 또는 sudo systemctl status sshd

# 3. 관련 명령어
서비스 중지: sudo systemctl stop ssh (또는 sshd)

서비스 시작: sudo systemctl start ssh (또는 sshd)

참고: systemctl 명령이 작동하지 않는 오래된 시스템에서는 service ssh restart 명령어를 사용할 수 있습니다.

# ufw allow 22/tcp 명령어
리눅스에서 SSH(22번 포트) 접속을 허용하기 위해 UFW(Uncomplicated Firewall)에서

sudo ufw allow 22/tcp 명령어를 사용합니다.

이는 TCP 프로토콜을 사용하는 22번 포트의 들어오는 연결만 허용하는

가장 일반적인 SSH 보안 설정입니다.
티스토리

# 1. 주요 명령어
SSH (TCP 22번) 허용: sudo ufw allow 22/tcp

SSH (서비스명으로 허용): sudo ufw allow ssh (22/tcp와 동일)

방화벽 설정 반영 및 확인: sudo ufw reload 후 sudo ufw status 명령어로 22/tcp ALLOW 항목을 확인.
# 2. 관련 명령어
IP 제한 허용: 특정 IP에서만 SSH 접속 허용 시 사용합니다.

sudo ufw allow from 192.168.0.1 to any port 22 proto tcp

규칙 삭제: 허용했던 22번 포트를 다시 차단할 때 사용합니다.

sudo ufw delete allow 22/tcp
# 3. 주의사항
UFW 활성화 필수: 방화벽이 작동 중이어야 규칙이 적용됩니다. (sudo ufw enable)

SSH 포트 변경 시: 만약 SSH 포트를 22번이 아닌 다른 포트(예: 2222)로 변경했다면 sudo ufw allow 2222/tcp로 설정해야 합니다.

# ssh ubuntu@192.168.111.100 명령어
리눅스 터미널(또는 Mac 터미널, Windows PowerShell/CMD)에서

192.168.111.100 IP를 가진 우분투 서버에 ubuntu 계정으로 SSH 접속을 시도하는 명령어는 다음과 같습니다.
# 1. 기본 접속 명령어

ssh ubuntu@192.168.111.100

이 명령어를 입력하면 ubuntu 사용자의 비밀번호를 묻는 프롬프트가 나타납니다.
# 2. 처음 접속하는 경우
처음 해당 서버에 접속하면 아래와 같은 메시지가 나옵니다.

The authenticity of host '192.168.111.100' can't be established.

... Are you sure you want to continue connecting (yes/no/[fingerprint])?

yes 를 입력하고 Enter를 누르면 접속이 진행됩니다.
# 3. 포트가 다른 경우 (22번이 아닐 때)
SSH 기본 포트인 22번이 아닌 다른 포트를 사용하는 경우

-p 옵션을 사용합니다. (예: 2222포트)

ssh -p 2222 ubuntu@192.168.111.100
# 4. 접속 문제 해결 (연결이 안 될 때)

* Samba에서 새로 생성되는 디렉터리(폴더)의 기본 권한을 설정하는 옵션

* directory mask는 디렉터리 생성 시 허용할 최대 권한을 지정

* 0777은 모든 사용자에게 읽기(r), 쓰기(w), 실행(x) 권한을 허용

* 실제 적용 권한은 시스템 umask와 함께 결정됨

* 협업/테스트 환경에서는 편리하지만, 운영 환경에서는 보안상 제한 권장

#   path = /share 명령어
* Samba에서 공유할 실제 디렉터리 경로를 지정하는 옵션
* path = /share는 서버의 /share 폴더를 네트워크 공유 대상으로 설정
* 클라이언트는 이 경로를 통해 해당 디렉터리의 파일에 접근
* smb.conf의 공유 섹션([share] 등) 안에 작성
* 실제 존재하는 디렉터리여야 정상적으로 공유 가능

#    read only = No 명령어
* Samba 공유 폴더의 읽기 전용 여부를 설정하는 옵션
* read only = No는 읽기와 쓰기 모두 허용 (쓰기 가능)
* 기본값은 보통 Yes(읽기 전용)로 설정되는 경우가 많음
* No로 설정하면 파일 생성, 수정, 삭제가 가능해짐
* 협업용 공유 폴더나 업로드 폴더에서 주로 사용

# valid users = @sambaGroup 명령어
* Samba 공유 폴더에 접근할 수 있는 사용자를 제한하는 옵션
* valid users = @sambaGroup은 sambaGroup 그룹에 속한 사용자만 접근 허용
* @ 기호는 특정 “그룹”을 의미
* 그룹에 속하지 않은 사용자는 해당 공유에 접근 불가
* 보안을 위해 공유 접근 범위를 제한할 때 사용

# systemctl enable smbd 명령어
* Samba 파일 공유 데몬(smbd)을 부팅 시 자동 실행되도록 설정
* enable은 systemd에 서비스 자동 시작을 등록하는 명령
* 서버 재부팅 후에도 Windows 파일 공유 기능이 자동으로 동작
* Samba 서버 운영 시 반드시 설정하는 기본 항목

# systemctl restart smbd 명령어
* Samba 파일 공유 데몬(smbd)을 재시작하는 명령
* smb.conf 설정 변경 후 적용할 때 사용
* stop(중지) 후 start(시작)를 한 번에 수행
* 공유 폴더 설정이나 권한 변경 후 반드시 실행 필요
* Windows 클라이언트 접속 문제 발생 시 기본 점검 단계

# systemctl status smbd 명령어
* Samba 파일 공유 데몬(smbd) 상태를 확인하는 명령
* 서비스가 정상적으로 실행 중인지(active / running) 여부를 확인
* failed 상태일 경우 설정 오류(smb.conf), 권한 문제, 포트 충돌 등을 의심
* Windows 클라이언트 접속 문제 해결 시 가장 기본적인 진단 단계

# ufw allow Samba 명령어
* UFW 방화벽에서 Samba 파일 공유 서비스 허용 규칙 추가
* Samba는 Windows 파일 공유(SMB/CIFS)를 위해 여러 포트(137~139, 445)를 사용
* "Samba"라는 서비스 이름으로 등록된 방화벽 프로필을 한 번에 허용
* Windows 클라이언트에서 공유 폴더 접근이 가능해짐
* 방화벽이 활성화된 경우 반드시 필요
