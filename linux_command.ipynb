# ls 명령어 설명
ls 명령어는 유닉스 계열 운영 체제(Linux 등)에서 현재 디렉토리 또는 지정한 디렉토리 내의 파일과 디렉토리 목록을 표시하는 필수 명령어입니다.

list segments의 약자로, 파일명, 디렉토리 구조, 그리고 -l 옵션을 통한 권한, 소유자, 크기 등 상세 정보를 확인하는 데 사용됩니다.

# 자주 사용하는 ls 옵션
ls -l: 파일의 상세 정보(권한, 소유자, 크기, 수정 시간)를 나열합니다.

ls -a: 숨김 파일(점.으로 시작하는 파일)을 포함한 모든 파일을 표시합니다.

ls -al: 상세 정보와 함께 숨김 파일을 모두 보여줍니다.

ls -h: -l 옵션과 함께 사용하여 파일 크기를 읽기 쉬운 단위(KB, MB, GB)로 표시합니다.

ls -t: 파일을 수정 시간 순서대로 최신순으로 정렬하여 표시합니다.

ls -r: 목록을 역순으로 정렬합니다.

ls -R: 하위 디렉토리의 내용까지 재귀적으로 모두 나열합니다.

ls -d: 디렉토리 자체의 정보를 보여줍니다.

# 활용 예시
ls -alF: 모든 파일을 상세 정보와 함께 표시하고, 파일 형식을 알리는 기호(*, / 등)를 끝에 붙입니다.

ls -ltR: 하위 디렉토리를 포함하여 시간순으로 정렬된 목록을 상세히 출력합니다.

# CD 명령어
리눅스의 cd(Change Directory) 명령어는 현재 작업 중인 디렉토리(폴더)를 이동할 때 사용하는 가장 기초적인 명령어입니다.

절대 경로 또는 상대 경로를 사용하여 디렉토리를 변경할 수 있으며,

cd ..으로 상위 폴더로, cd 또는 cd ~으로 홈 디렉토리로 바로 이동할 수 있습니다.
# 주요 cd 명령어 사용법:
cd [디렉토리명]: 현재 위치 내의 특정 디렉토리로 이동

cd ..: 상위 디렉토리로 한 단계 이동

cd /: 루트(/) 디렉토리로 이동

cd ~ 또는 cd: 현재 사용자의 홈 디렉토리로 이동

cd -: 바로 이전에 위치했던 디렉토리로 이동

cd /home/user/docs: 절대 경로를 사용하여 특정 디렉토리로 직접 이동

팁:
디렉토리 이름을 입력할 때 Tab 키를 사용하면 자동 완성을 통해 더 편하게 이동할 수 있습니다.

현재 위치는 pwd 명령어로 확인 가능합니다.

# PWD 명령어
리눅스 pwd (Print Working Directory) 명령어는 사용자가 현재 작업 중인 디렉토리의 전체 절대 경로를 화면에 출력합니다.

터미널에서 위치를 확인하거나 스크립트 작성 시 현재 위치를 파악하는 데 필수적이며, 주로 cd (디렉토리 이동) 명령과 함께 사용됩니다.
# 주요 내용:
용도: 현재 위치(Working Directory) 확인.

사용법: 터미널에 pwd 입력.

특징: 최상위 루트(/)부터 현재 위치까지의 절대 경로를 보여줌.

# 옵션:
pwd -P: 물리적 경로(심볼릭 링크가 아닌 실제 경로) 표시.

pwd -L: 논리적 경로(심볼릭 링크 포함) 표시.

복잡한 디렉토리 구조에서 작업할 때 자신이 어디에 있는지 명확히 알고자 할 때 매우 유용합니다.

# touch 명령어
리눅스의 touch 명령어는 0바이트 빈 파일을 생성하거나,

기존 파일의 최근 접근/수정 시간(atime, mtime)을 현재 시간으로 갱신하는 데 사용됩니다.

파일이 없으면 새로 만들고, 있으면 시간만 변경하며 touch [파일명] 형태로 사용합니다.

# 주요 기능 및 사용법
빈 파일 생성: touch newfile.txt (파일이 없으면 생성)

시간 갱신: touch existingfile.txt (이미 존재하는 파일의 시간 정보를 현재로 업데이트)

여러 파일 생성: touch file1.txt file2.txt (공백으로 구분하여 동시에 여러 파일 생성)

# 주요 옵션
-c: 파일이 없어도 새로 생성하지 않습니다. (시간 업데이트만 수행)

-a: 최근 접근 시간(access time)만 변경합니다.

-m: 최근 변경 시간(modification time)만 변경합니다.

-t: [[CC]YY]MMDDhhmm[.ss] 형식으로 시간을 직접 지정하여 변경합니다.

# rm 명령어
리눅스의 rm (remove) 명령어는 파일이나 디렉토리를 삭제하는 핵심 명령어입니다.

기본적으로 파일을 삭제하며, -r (디렉토리 삭제), -f (강제 삭제) 등의 옵션을 사용하여

하위 디렉토리를 포함한 폴더 전체를 강제로 지울 수 있습니다.

삭제된 파일은 복구가 어렵기 때문에 주의가 필요합니다.
# 주요 사용법 및 옵션
파일 삭제: rm [파일명] (기본적으로 삭제 여부 확인)

-i (Interactively): 삭제 전 물어봄 (rm -i file.txt)

-r (Recursive): 디렉토리 및 내부 파일 전체 삭제 (rm -r folder)

-f (Force): 존재하지 않는 파일은 무시하고, 묻지 않고 강제 삭제 (rm -f file.txt)

-rf: 디렉토리를 묻지 않고 강제로 삭제 (rm -rf folder)

# cp 명령어
리눅스 cp (copy) 명령어는 파일이나 디렉터리를 복사할 때 사용하며, 기본 구문은 cp [옵션] [원본] [대상]입니다.

파일 복사, 이름 변경 복사, 디렉터리 복사(-r), 덮어쓰기 확인(-i), 권한 유지(-p) 등

다양한 옵션을 지원하여 시스템 관리에서 필수적인 명령어입니다.
# 주요 cp 명령어 사용법 및 예제
파일 복사: cp file1.txt file1_bak.txt (file1을 file1_bak로 복사)

파일을 디렉터리로 복사: cp file1.txt /home/user/documents/

디렉터리 복사 (하위 포함): cp -r dir1 dir2 (-r 옵션 필수)

복사 시 덮어쓰기 확인: cp -i file1.txt file2.txt (대상 파일 존재 시 물어봄)

파일 속성(권한) 유지하며 복사: cp -p file1.txt file1_bak.txt

여러 파일 복사: cp file1.txt file2.txt /dest/dir/

# 자주 사용하는 핵심 옵션
-r, -R: 디렉터리를 복사할 때 하위 디렉터리와 파일을 모두 복사

-i: 복사할 목적지에 동일한 이름의 파일이 존재할 경우, 덮어쓸지 물어봄

-f: 목적지에 파일이 있어도 강제로 덮어씀

-p: 원본 파일의 권한, 소유자, 시간 등 속성을 보존

-v: 복사 과정을 화면에 표시 (verbose)

# mv 명령어
리눅스 mv (move) 명령어는 파일이나 디렉토리를 다른 위치로 이동하거나 이름을 변경할 때 사용합니다.

기본 문법은 mv [옵션] <원본> <대상>이며, 같은 디렉토리 내에서 대상을 다르게 지정하면 이름이 변경되고,

다른 디렉토리 경로를 지정하면 파일이 이동됩니다.
# 주요 활용법 및 명령어:
파일 이름 변경: mv old_name.txt new_name.txt

파일 위치 이동: mv file.txt /path/to/destination/

디렉토리 이동: mv /source_dir /path/to/destination/

여러 파일 이동: mv file1.txt file2.txt /dest_dir/

# 주요 옵션:
-i (interactive): 덮어쓰기 전 확인 메시지를 표시

-f (force): 덮어쓰기 확인 없이 강제 이동

-n (no-clobber): 대상 파일이 존재하면 덮어쓰지 않음

-v (verbose): 이동 중인 파일 목록을 상세히 출력

# 주의사항:
동일한 이름의 파일이 목적지에 있으면 경고 없이 덮어씌워지므로 주의가 필요합니다.

-i 옵션을 사용하여 덮어쓰기 전에 확인하는 것이 안전합니다.

# mkdir 명령어
리눅스 mkdir(make directory)은 새로운 디렉토리(폴더)를 생성하는 명령어입니다.

mkdir [옵션] 디렉토리명 형식을 사용하며, -p 옵션으로 상위 디렉토리까지 한번에 생성하거나, -m으로 권한을 설정할 수 있습니다.

# 주요 사용법 및 옵션
기본 생성: mkdir folder1 (현재 위치에 folder1 생성)

여러 디렉토리 생성: mkdir dir1 dir2 dir3 (공백으로 구분)

하위 디렉토리 포함 생성 (-p): mkdir -p parent/child (상위 디렉토리 없어도 자동 생성)

권한 지정하여 생성 (-m): mkdir -m 755 new_folder (생성과 동시에 권한 설정)

생성 내용 출력 (-v): mkdir -v new_folder (디렉토리 생성 메시지 출력)

# 주의사항:
동일한 이름의 디렉토리가 있으면 에러가 발생하므로, 기존 디렉토리가 있어도 에러 없이 진행하려면 -p 옵션을 활용합니다.
